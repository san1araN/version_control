Условия
Выгрузить игроков, которые:     
С 12:00 14.03.2022 сделали ставку в пре-матч на события раздела «Киберспорт».
Минимальная сумма ставки для участия в Бонусном предложении составляет 10 BYN с коэффициентом не менее 1,5, в экспрессе все события на киберспорт с кф каждого события от 1.50
Ставка должна быть рассчитана не позднее 12:00 15.03.2022.
Ставки вида «система» не учитываются.
Ставки CashOut, возвраты и FreeBet не участвуют в Бонусе.

Решение необходимо в двух вариантах: 
SQL 
Python


```python
import pandas as pd
bets = pd.read_csv('E:\Data Analyst\\tz\Betera\\bets.csv')
events = pd.read_csv('E:\Data Analyst\\tz\Betera\\events.csv')
```

```python
# преобразование в тип данных datetime
bets['accept_time'] = pd.to_datetime(bets['accept_time'])
bets['settlement_time'] = pd.to_datetime(bets['settlement_time'])
```

```python
# объеденим две таблицы
be = pd.merge(bets,events, on='event_id', how='left')
```

```python
# фильтрация по дате, времени, ивенту, спорту, типу ставки, результату ставки и коэффу
data = be[(be['accept_time'] >= '2022-03-14 12:00')&(be['settlement_time'] <= '2022-03-15 12:00')
          &(be['event_stage'] == 'Prematch')&(be['sport'] == 'E-Sports')&(be['bet_type']!='System')
          &(be['result']!='Cashout')&(be['result']!='Return')&(be['is_free_bet']!='True')&(be['accepted_bet_odd']>=1.5)]
```

```python
# фильтрация по минимальной сумме ставки в ординаре от 10
df = data[(data['bet_type'] != 'Ordinar')|(data['amount']>=10)]
df
```
